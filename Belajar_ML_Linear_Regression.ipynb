```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

# Generate data random

```python
jumlah_data = 100

y = np.array([i*0.1+np.random.randn() for i in range(jumlah_data)]) #menghasilkan 100 data di list, dengan sedikit simpangan (rand()), untuk memperbesar sebarannya kita bisa buat valuenya sampai 10 saja (0,1), pakai numpy array biar pemrosesannya lebih cepat
x = np.array([i*0.1 for i in range(jumlah_data)])
```

## 1. Plot data

```python
plt.scatter(x, y)
plt.show()
```

## 2. Fungsi linear

```python
def fungsiLinear(x, gradien):
  y = gradien*x
  return y
```

Menghitung persamaan linear awal

```python
x_prediction = np.array([0,10])
m_init = 5
y_prediction = fungsiLinear(x_prediction, m_init)
```

```python
plt.scatter(x,y)
plt.plot(x_prediction, y_prediction, "Red")
plt.show()
```

```python
plt.scatter(x,y)
plt.plot(x_prediction, y_prediction, "Red")
plt.axis([0, 10, 0, 10])
plt.show()
```

Menghitung Prediksi

```python
m_prediction = m_init
m_list_prediction = []
y_list_prediction = []
x_list_prediction = []
learning_rate = 0.1


# Secara matematis, persamaan garis regresi linear dapat dituliskan sebagai:

# 𝑦 = 𝑚𝑥+𝑐 atau y = m.x


for i in range(1, jumlah_data): #awali dengan 1 karena data pertama (index pertama) = 0
  y_prediction = fungsiLinear(x[i], m_prediction)
  y_actual = y[i]

  error = y_actual-y_prediction

  delta_m = learning_rate*error/x[i]
  m_prediction = m_prediction + delta_m #update prediksi agar makin presisi dengan nilai actualnya (agar errornya mendekati 0)

  m_list_prediction.append(m_prediction)
  y_list_prediction.append(fungsiLinear(np.array([0,10]), m_prediction))
  x_list_prediction.append(np.array([0,10]))
```

```python
print(y_prediction)
```

Visualisasi Prediksi

```python
fig = plt.figure(figsize=(4,4))
line, = plt.plot([],[],"Red")

def animate(frame_num):
	x = x_list_prediction[frame_num]
	y = y_list_prediction[frame_num]
	line.set_data((x,y))
	return line

plt.scatter(x,y)
plt.axis([0,10,0,10])
anime = FuncAnimation(fig,animate,frames=100,interval=100,repeat=False)
plt.show()
```
